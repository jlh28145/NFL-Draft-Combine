```python
import pandas as pd
import nfl_data_py as nfl

years = range(2014,2024)
df = nfl.import_combine_data(years)
```

```python
df.groupby('draft_year').count()
```

```python
# Display the shape of the dataset (rows, columns)
print("Dataset shape:", df.shape)

# Show the first 5 rows of the dataset to get an overview
print(df.head())

# List all column names and their data types
print("\nColumn Info:")
print(df.dtypes)

# Get a basic statistical summary of numerical columns
print("\nStatistical Summary:")
print(df.describe())

# Check for missing values in each column
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ðŸ“Š 1. Summary Statistics
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# General info about the dataset
df.info()

# Summary statistics for numeric columns
df.describe().T
```

```python
# ðŸ“ˆ 2. Value Counts & Distribution Plots

# Set plot style
sns.set(style="whitegrid")

# Draft round frequency
plt.figure(figsize=(8, 4))
sns.countplot(x='draft_round', data=df)
plt.title('Distribution of Draft Rounds')
plt.xlabel('Draft Round')
plt.ylabel('Player Count')
plt.show()

# Position frequency
plt.figure(figsize=(12, 4))
sns.countplot(x='pos', data=df, order=df['pos'].value_counts().index)
plt.title('Distribution of Positions')
plt.xlabel('Position')
plt.ylabel('Player Count')
plt.xticks(rotation=45)
plt.show()

# Distribution of 40-yard dash times
plt.figure(figsize=(8, 4))
sns.histplot(df['forty'], bins=30, kde=True)
plt.title('Distribution of 40-Yard Dash Times')
plt.xlabel('Time (seconds)')
plt.ylabel('Number of Players')
plt.show()
```

```python
# ðŸ”¥ 3. Correlation Heatmap

# Select only numeric columns
numeric_cols = df.select_dtypes(include='number')

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
# Show total missing values per column
missing_counts = df.isnull().sum().sort_values(ascending=False)
print("Missing Values Per Column:\n", missing_counts)

# Show percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing Percentage:\n", missing_percentage.sort_values(ascending=False))
```

```python
# â“ 4. Missing Value Heatmap & Bar Plot

# Missing value heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()

# Bar plot of missing values per column
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]

plt.figure(figsize=(10, 5))
missing_counts.sort_values(ascending=False).plot(kind='bar')
plt.title('Missing Values per Column')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45)
plt.show()
```
