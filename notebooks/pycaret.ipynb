```python
# Feature Selection Notebook

import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
# from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

# Load cleaned and preprocessed dataset
df = pd.read_csv("../data/processed/combine_preprocessed.csv")

# Define targets
target_classification = df['draft_round']
target_regression = df['draft_ovr']
```

```python
from pycaret.classification import *

# Create a PyCaret classification setup
clf_setup = setup(
    data=df, 
    target='draft_round', 
    # categorical_features=['Pos', 'School'],  # if not already encoded
    session_id=42,
    train_size=0.8,
    preprocess=True,
    #silent=True,
    verbose=False
)

# Compare models
best_class_model = compare_models(include=['lr', 'rf', 'dt', 'knn', 'nb', 'gbc'], turbo=True)

# Finalize and save
final_class_model = finalize_model(best_class_model)
save_model(final_class_model, './../models/pycaret_classification_model')
```

```python
from pycaret.regression import *

# Create a PyCaret regression setup
reg_setup = setup(
    data=df,
    target='draft_ovr',
    #categorical_features=['Pos', 'School'],
    session_id=42,
    train_size=0.8,
    preprocess=True,
    #silent=True,
    verbose=False
)

# Compare models
best_reg_model = compare_models(include=['lr', 'rf', 'dt', 'knn', 'ada', 'gbr', 'lasso'], turbo=True)

# Finalize and save
final_reg_model = finalize_model(best_reg_model)
save_model(final_reg_model, './../models/pycaret_regression_model')
```
